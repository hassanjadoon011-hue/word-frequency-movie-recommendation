```python
# Cell 1: Setup
import os
import re
import time
import random
from collections import Counter
import matplotlib.pyplot as plt

print("Word Frequency Analyzer")
print("System ready")
```

```python
# Cell 2: Create files
os.makedirs("text_files", exist_ok=True)

files = [
    ("file1.txt", "apple banana apple orange banana grape apple orange mango"),
    ("file2.txt", "car bus car train bike car bus train plane car bike"),
    ("file3.txt", "cat dog cat bird fish dog cat rabbit bird fish cat"),
]

for filename, content in files:
    with open(f"text_files/{filename}", "w") as f:
        f.write(content)
    print(f"Created: {filename}")

print("Files created")
```

```python
# Cell 3: Word Counter
class WordCounter:
    def __init__(self):
        self.stop_words = {'a', 'an', 'the', 'and', 'or', 'but'}

    def count_words(self, text):
        if not text.strip():
            return {'total': 0, 'unique': 0, 'counts': {}}

        text = text.lower()
        words = re.findall(r'\b[a-z]{2,}\b', text)
        filtered = [w for w in words if w not in self.stop_words]
        word_counts = Counter(filtered)

        return {
            'total': len(filtered),
            'unique': len(word_counts),
            'counts': dict(word_counts)
        }

    def show_results(self, result):
        print(f"\nTotal words: {result['total']}")
        print(f"Unique words: {result['unique']}")

        if result['counts']:
            print("\nWord counts:")
            for word, count in result['counts'].items():
                print(f"{word}: {count}")

counter = WordCounter()
print("Word counter ready")
```

```python
# Cell 4: Movie Database
class MovieDatabase:
    def __init__(self):
        self.movies = {
            "boxing": ["Rocky", "Creed", "Million Dollar Baby", "Raging Bull"],
            "action": ["John Wick", "Mission Impossible", "Die Hard", "Gladiator"],
            "horror": ["The Shining", "Get Out", "The Conjuring", "It"],
            "comedy": ["Superbad", "The Hangover", "Step Brothers", "Bridesmaids"],
            "drama": ["The Shawshank Redemption", "Forrest Gump", "The Godfather", "Parasite"]
        }

    def get_movies(self, category):
        category = category.lower().strip()

        for key in self.movies.keys():
            if key in category or category in key:
                return self.movies[key]

        return []

    def show_movies(self, category, movies_list):
        if not movies_list:
            print(f"\nNo movies found for '{category}'")
            return

        print(f"\nMovies for {category}:")
        for i, movie in enumerate(movies_list, 1):
            print(f"{i}. {movie}")

movie_db = MovieDatabase()
print("Movie database ready")
```

```python
# Cell 5: Performance Tester (FIXED)
class PerformanceTester:
    def test_performance(self, counter):
        print("\nRunning performance test...")

        # Test text
        test_text = " ".join(["word"] * 1000)

        # Serial
        start = time.time()
        counter.count_words(test_text)
        serial_time = time.time() - start

        # Parallel simulation
        start = time.time()
        words = test_text.split()
        chunk_size = len(words) // 4

        for i in range(4):
            start_idx = i * chunk_size
            end_idx = start_idx + chunk_size if i < 3 else len(words)
            chunk_words = words[start_idx:end_idx]
            chunk_text = " ".join(chunk_words)
            counter.count_words(chunk_text)

        parallel_time = time.time() - start

        speedup = serial_time / parallel_time if parallel_time > 0 else 1

        # Create graph
        plt.figure(figsize=(8, 5))
        bars = plt.bar(['Serial', 'Parallel'], [serial_time, parallel_time], color=['blue', 'green'])
        plt.title('Performance Comparison')
        plt.ylabel('Time (seconds)')

        # Add values on bars
        for bar, t in zip(bars, [serial_time, parallel_time]):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.00001,
                    f'{t:.6f}s', ha='center', va='bottom')

        plt.show(block=False)  # Don't block execution
        plt.pause(0.1)  # Small pause

        print(f"\nResults:")
        print(f"Serial time: {serial_time:.6f}s")
        print(f"Parallel time: {parallel_time:.6f}s")
        print(f"Speedup: {speedup:.2f}x")

        # Close the figure to continue
        plt.close()

        return serial_time, parallel_time, speedup

tester = PerformanceTester()
print("Performance tester ready")
```

```python
# Cell 6: Main Application (FIXED - will continue after graph)
def main():
    print("\n" + "="*40)
    print("WORD FREQUENCY ANALYZER")
    print("="*40)

    while True:
        print("\nMAIN MENU")
        print("1. Count words")
        print("2. Analyze file")
        print("3. Get movies")
        print("4. Performance test")
        print("5. Exit")
        print("-" * 20)

        choice = input("Choice (1-5): ").strip()

        if choice == '1':
            print("\nEnter text:")
            text = input("Text: ").strip()

            if not text:
                text = "hello world hello test"
                print("Using: hello world hello test")

            result = counter.count_words(text)
            counter.show_results(result)

        elif choice == '2':
            print("\nAvailable files:")
            files = os.listdir("text_files")
            for i, f in enumerate(files, 1):
                print(f"{i}. {f}")

            try:
                file_num = input("\nFile number: ").strip()
                if file_num.isdigit():
                    idx = int(file_num) - 1
                    if 0 <= idx < len(files):
                        filename = files[idx]
                        with open(f"text_files/{filename}", "r") as f:
                            text = f.read()

                        print(f"\nFile: {filename}")
                        result = counter.count_words(text)
                        counter.show_results(result)
                    else:
                        print("Invalid number")
                else:
                    print("Enter a number")
            except Exception as e:
                print(f"Error: {e}")

        elif choice == '3':
            print("\nCategories: boxing, action, horror, comedy, drama")
            category = input("Enter category: ").strip().lower()

            if not category:
                category = "boxing"
                print("Using: boxing")

            movies = movie_db.get_movies(category)
            movie_db.show_movies(category, movies)

        elif choice == '4':
            print("\nRunning performance test...")
            serial, parallel, speedup = tester.test_performance(counter)

            # Now it will continue here after graph
            print("\nTest completed. Returning to menu...")

        elif choice == '5':
            print("\nGoodbye!")
            break

        else:
            print("Invalid choice")

# Run the application
if __name__ == "__main__":
    main()
```
